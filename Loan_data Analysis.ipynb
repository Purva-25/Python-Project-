## Null Treatment 

```python
df.isna().sum()
```

## Null treatment

```python
df.isna().sum().plot(kind='bar')
```

```python
for i in df.columns:
    if df[i].dtypes==object:
        df[i]=df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].mean())
```

```python
df.isna().sum().plot(kind='bar')
```

## Descriptive Statics

```python
for i in df.columns:
    if df[i].dtypes!=object:
        print(i)
        print()
        print(df[i].describe())
        print('******************************************************')
```
